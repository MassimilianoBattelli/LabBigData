```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
df
```

```python
df = pd.DataFrame(df['Energy;Date;Topic'].str.split(';',).tolist(),columns = ['Energy','Date','Topic'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], utc = True).dt.tz_convert('Europe/Berlin')
```

```python
df = df[['Energy', 'Date']]
```

```python
df
```

```python
df = df.sort_values(['Date'])
```

```python
df['Energy'] = df['Energy'].astype('float')
```

```python
df['Consumo'] = (df['Energy'] - df['Energy'].shift(1))
```

```python
df
```

```python
df = df[df.Consumo >= 0]
```

```python
df
```

```python
df = df.resample('H', on='Date').agg({'Consumo': 'sum'}).reset_index()
```

```python
df
```

```python
df['DayOfWeek'] = df['Date'].dt.dayofweek
```

```python
df['Hour'] = df['Date'].dt.hour
```

```python
df = df.drop(columns=['Date'])
```

```python
df
```

```python
df = df.groupby(['DayOfWeek', 'Hour']).mean()
```

```python
df['Consumo'] = df['Consumo']/1000
```

```python
df = df.unstack()
```

```python
df.columns = df.columns.droplevel(0)
```

```python
df
```

```python
import seaborn as sns
s = sns.heatmap(df, linewidths=.5)
```
