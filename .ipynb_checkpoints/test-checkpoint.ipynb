```python
import seaborn as sns
from datetime import datetime
import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import statsmodels.api as sm

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, plot_predict # per plottare ACF, PACF e previsione

from statsmodels.tsa.stattools import adfuller # per il test di verifica della stazionarietà della serie storica

from statsmodels.tsa.arima.model import ARIMA # per la stima del modello ARIMA
```

```python
df = pd.read_csv('data.csv', delimiter=";")
```

```python
df
```

```python
df.info()
```

**Divido il dataframe in 3 colonne**

```python
df['Date'] = pd.to_datetime(df['Date'], utc = True).dt.tz_convert('Europe/Berlin')
```

```python
df.info()
```

**Per l'analisi non ho bisogno della colonna 'Topic' perciò la elimino**

```python
df = df.sort_values(['Date'])
```

**Creo la colonna 'Consumo' che indicherà quanta energia è stata utilizzata dalla misurazione precedente a quella attuale**

```python
df['Energy'] = df['Energy'].astype('float')
```

```python
df['Consumo'] = df['Energy'].diff()
```

```python
df = df.drop(columns=['Topic', 'Energy'])
```

```python
df = df.set_index("Date")
```

```python
df.reset_index()
```

**Può succedere che il sensore si spenga resettando il valore di energia misurato fino a quel momento,
  questo causa dei valori negativi sballati nel Consumo**

```python
plot = df["Consumo"].plot()
plot.set_xlabel("Time")
plot.set_ylabel("Consumo")
```

**Elimino i valori sballati prendendo solo quelli positivi**

```python
df = df[df['Consumo']>= 0]
```

```python
plot = df["Consumo"].plot()
plot.set_xlabel("Time")
plot.set_ylabel("Consumo")
```

```python
df = df.reset_index()
df = df.resample('H', on='Date').agg({'Consumo': 'sum'})
```

```python
plot = df["Consumo"].plot()
plot.set_xlabel("Time")
plot.set_ylabel("Consumo")
```

```python
plot_acf(df["Consumo"])
```

```python
result = adfuller(df["Consumo"])
result
```

```python
print('ADF Statistic: {}'.format(result[0]))
print('Level of significance: {}'.format(result[4]))
print('p-value: {}'.format(result[1]))
```

```python
df = df.reset_index()
```

```python
df = df.resample('D', on='Date').agg({'Consumo': 'sum'}).reset_index()
```

```python
df['date_delta'] = (df['Date'] - df['Date'].min())  / np.timedelta64(1,'D')
```

```python
x = df['date_delta']
y = df['Consumo']
```

```python
plt.scatter(x,y)
plt.xlabel('Time')
plt.ylabel('Consumption')
plt.show()
```

```python
np.asarray(x)
```

```python
x1 = sm.add_constant(x)
results = sm.OLS(y,x1).fit()
results.summary()
```

```python
plt.scatter(x,y)
yhat = x*77.6284+5448.8591
fig = plt.plot(x, yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('Time')
plt.ylabel('Consumption')
plt.show()
```

**Ricampiono la 'Date' per intervalli orari, sommando i consumi**

```python
df = df.resample('H', on='Date').agg({'Consumo': 'sum'}).reset_index()
```

```python
df
```

**Creo 2 colonne: Ora, e Giorno della settimana**

```python
df['DayOfWeek'] = df['Date'].dt.dayofweek
```

```python
df['Hour'] = df['Date'].dt.hour
```

```python
df = df.drop(columns=['Date'])
```

```python
df
```

**Raggruppo i valori delle ore per giorno della settimana, facendo la media,
  divido il Consumo per 1000 per avere il valore in kW**

```python
df = df.groupby(['DayOfWeek', 'Hour']).mean()
```

```python
df['Consumo'] = df['Consumo']/1000
```

```python
df
```

```python
df = df.unstack()
```

```python
df.columns = df.columns.droplevel(0)
```

```python
df
```

```python
s = sns.heatmap(df, cmap="viridis", cbar_kws={'label': 'kW'}, linewidths=.5, linecolor = 'black')
```
