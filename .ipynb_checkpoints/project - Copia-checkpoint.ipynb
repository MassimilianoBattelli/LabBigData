```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

**Divido il dataframe in 3 colonne**

```python
df = pd.DataFrame(df['Energy;Date;Topic'].str.split(';',).tolist(),columns = ['Energy','Date','Topic'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], utc = True).dt.tz_convert('Europe/Berlin')
```

```python
df
```

**Per l'analisi non ho bisogno della colonna 'Topic' perciò la elimino**

```python
df = df[['Energy', 'Date']]
```

```python
df = df.sort_values(['Date'])
```

**Creo la colonna 'Consumo' che indicherà quanta energia è stata utilizzata dalla misurazione precedente a quella attuale**

```python
df['Energy'] = df['Energy'].astype('float')
```

```python
df['Consumo'] = df['Energy'].diff()
```

```python
df
```

**Può succedere che il sensore si spenga resettando il valore di energia misurato fino a quel momento,
  questo causa dei valori negativi sballati nel Consumo**

```python
df.plot(x = 'Date', y = "Consumo")
plt.show()
```

**Elimino i valori sballati prendendo solo quelli positivi**

```python
df = df[df['Consumo']>= 0]
```

```python
df.plot(x = 'Date', y = "Consumo")
plt.show()
```

**Creo due DF ricampionando la 'Date' per intervalli giornalieri e orari, sommando i consumi**

```python
dfMonth = df.resample('D', on='Date').agg({'Consumo': 'sum'}).reset_index()
dfMonth
```

```python
dfWeek = df.resample('H', on='Date').agg({'Consumo': 'sum'}).reset_index()
dfWeek
```

**Creo 2 colonne rispettivamente: Mese-Giorno e Giorno-Ora della settimana**

```python
dfMonth['Month'] = dfMonth['Date'].dt.month
dfMonth['Day'] = dfMonth['Date'].dt.day
dfMonth = dfMonth.drop(columns=['Date'])
dfMonth
```

```python
dfWeek['DayOfWeek'] = dfWeek['Date'].dt.dayofweek
dfWeek['Hour'] = dfWeek['Date'].dt.hour
dfWeek = dfWeek.drop(columns=['Date'])
dfWeek
```

**Raggruppo i valori facendo la media,
  divido il Consumo per 1000 per avere il valore in kW**

```python
dfMonth = dfMonth.groupby(['Month', 'Day']).mean()
dfMonth['Consumo'] = dfMonth['Consumo']/1000
dfMonth
```

```python
dfWeek = dfWeek.groupby(['DayOfWeek', 'Hour']).mean()
dfWeek['Consumo'] = dfWeek['Consumo']/1000
dfWeek
```

```python
dfMonth = dfMonth.unstack()
dfWeek = dfWeek.unstack()
```

```python
dfMonth.columns = dfMonth.columns.droplevel(0)
dfWeek.columns = dfWeek.columns.droplevel(0)
```
